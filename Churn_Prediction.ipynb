```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/drive/MyDrive/IIIT-A Internship/Telco-Customer-Churn.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(['customerID'], axis=1)
df.head()
```

```python
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce')
df.isnull().sum()
```

```python
df[np.isnan(df['TotalCharges'])]
```

```python
df.drop(df[df['tenure'] == 0].index, axis=0, inplace=True)
```

```python
df.head()
```

```python
df.fillna(df["TotalCharges"].mean())
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: "No", 1: "Yes"})
df.head()
```

```python
from sklearn.preprocessing import *

label_encoder = LabelEncoder()

colList = df.columns.tolist()

def obj_to_int(obj_series):
  if obj_series.dtype=='object':
    obj_series = label_encoder.fit_transform(obj_series)

  return obj_series

df = df.apply(lambda x: obj_to_int(x))
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
scaler= StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred_1 = knn.predict(X_test)
```

```python
from sklearn.metrics import *
print(classification_report(y_test, y_pred_1))
```

```python
confusion_matrix(y_test, y_pred_1)
```

# SVM Classifier

```python
from sklearn.svm import SVC

svc = SVC(random_state=1)
svc.fit(X_train, y_train)
```

```python
y_pred_2 = svc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_2))
```
