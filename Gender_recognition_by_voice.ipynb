```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import *

path = '/content/drive/MyDrive/IIIT-A Internship/voice.csv'
df = pd.read_csv(path)
df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['label'])
df['label'].unique()
```

```python
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
## Data Standardisation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# SVM Classifier

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(X_train, y_train)
```

```python
y_pred_1 = svc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_1))
```

```python
print("Accuracy: ", accuracy_score(y_test,y_pred_1)*100)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion="gini")
dt.fit(X_train, y_train)
```

```python
y_pred_2 = dt.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_2))
```

```python
print("Accuracy", accuracy_score(y_test, y_pred_2)*100)
```

# KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
y_pred_3 = knn.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_3))
```

```python
print("Accuracy", accuracy_score(y_test, y_pred_3)*100)
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_5 = gnb.predict(X_test)
print(classification_report(y_test, y_pred_5))
```
