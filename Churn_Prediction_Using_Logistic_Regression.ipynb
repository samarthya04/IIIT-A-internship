```python
from sklearn.preprocessing import *

label_encoder = LabelEncoder()

colList = df.columns.tolist()

def obj_to_int(obj_series):
  if obj_series.dtype=='object':
    obj_series = label_encoder.fit_transform(obj_series)

  return obj_series

df = df.apply(lambda x: obj_to_int(x))
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.values[:, -1]
```

```python
# y = pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
scaler= StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print("Shape of X_train", X_train.shape)
print("Shape of y_train", y_train.shape)
print("Shape of X_test", X_test.shape)
print("Shape of y_test", y_test.shape)
```

```python
# Reshape

X_train = X_train.T
y_train = y_train.reshape(1, X_train.shape[1])

X_test = X_test.T
y_test = y_test.reshape(1, X_test.shape[1])
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def LogisticRegressionModel(X, y, learning_rate=0.005, iterations=1000):

    n, m = X.shape

    W = np.zeros((n, 1))
    B = 0
    cost_list = []

    for i in range(iterations):
        A = sigmoid(np.dot(W.T, X) + B)

        cost = -(1/m)*np.sum(y*np.log(A) + (1-y)*np.log(1-A))

        dW = (1/m)*np.dot(A-y, X.T)
        dB = (1/m)*np.sum(A-y)

        W = W - learning_rate*dW.T
        B = B - learning_rate*dB

        cost_list.append(cost)

    return W, B, cost_list
```

```python
W, B, cost_list = LogisticRegressionModel(X_train, y_train)
```

```python
def accuracy(X, y, W, B):
    A = sigmoid(np.dot(W.T, X) + B)

    A = A > 0.5
    A = np.array(A, dtype='int64')

    acc = (1 - np.sum(np.absolute(A-y))/y.shape[1])*100

    print("Accuracy: ", acc)
```

```python
accuracy(X_test, y_test, W, B)
```
